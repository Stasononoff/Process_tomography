```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import linalg as sLA

from IPython.display import Image
from IPython.display import display, Latex

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt

from matplotlib import style
# style.use('ggplot')
```

```python
# Функция для отрисовки 3-D barchart
def plot_matrix(m, title = 'Входное состояние'):
    
    style.use('ggplot')

    fig = plt.figure(figsize = (8,8))
    ax1 = fig.add_subplot(111, projection='3d')
    
    xi = np.reshape(list(np.arange(m.shape[0]))*m.shape[1], (-1))
    yi = np.reshape(np.reshape((list(np.arange(m.shape[1]))*m.shape[0]), (m.shape[1], m.shape[0])).T, (-1))
    z0 = np.zeros(m.shape[0]*m.shape[1])
    
    dx = np.ones(m.shape[0]*m.shape[1])
    dy = np.ones(m.shape[1]*m.shape[0])
    dz = np.reshape(m, (-1))
    
    ax1.bar3d(xi, yi, z0, dx, dy, dz, shade=True)


    ax1.set_xlabel('Колличество фотонов во втором канале n2')
    ax1.set_ylabel('Колличество фотонов в первом канале n1')
    ax1.set_zlabel('Вероятность p1p2')
    ax1.set_title(title)

    plt.show()
```

Вариант 3:

Статистика фотонов Пуассоновская с $\mu = 0.18$:

$p(n) = \frac{\mu^{n}}{n!} \exp^{-\mu}$, где $\mu = 0.18$

# Пункт 1:

```python
# Пуассоновское распределение
def poisson_prob(n, mu = 0.18):
    return mu**n/np.math.factorial(n)*np.exp(-mu)
    
# Ошибки при рассмотрении заданного количества мод 
# заданного распределения
def restrictions_err(N, func = poisson_prob):
    return 1 - sum([func(i) for i in range(N)])
    
```

```python
print('Вероятность генирации Фоковского состояния 0:', poisson_prob(0))
print('Вероятность генирации Фоковского состояния 1:', poisson_prob(1))
```

```python
print('Ошибка в распределении за счёт сокращения размерности до 4:', restrictions_err(4))
print('Ошибка в распределении за счёт сокращения размерности до 5:', restrictions_err(5))
```

При размерности гильбертова пространства $>= 5$ Ошибка моделирования распределения < $10^{-5}$


# Пункт 2:

```python
# Производящая функция для заданного распределения и 
# заданного числа рассматриваемых мод (Численная оценка)
def G(z, N = 5, func = poisson_prob):
    return sum([z**n*func(n) for n in range(N)])
```

```python
print('Значение производящей фу-ии в нуле G(0): {}'.format(G(0)))
```

Когерентные состояния описываются статистикой Пуассона.
Известно, что для когерентных состояний аналитическое выражение
для производящей функцииможно записать в виде:

$G(Z) = e^{\mu (z-1)}$

```python
# Сравним численное значение G(Z,N) при размерности N=5 с аналитическим G(Z)

# Аналитическая формула
def G_poisson(z, mu = 0.18):
    return np.exp(mu*(z-1))

def G_real(z, func = G_poisson):
    return func(z)
    

z_list = np.linspace(-2,2,100)
G_real = G_real(z_list)
G_num = [G(z, N = 5) for z in z_list]

plt.title(r'Сравнение численной ооценки $G(\theta)$ с аналитическим значением')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$G(\theta)$')
plt.plot(z_list, G_real, linewidth = 3, linestyle = '-.', label = 'real')
plt.plot(z_list, G_num,  linestyle = '-', label = 'numerical')
plt.legend()
# plt.grid()
plt.show()

plt.title(r'Ошибка численной ооценки $G(\theta)$')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$Err(\theta)$')
plt.plot(z_list, G_real - G_num, linewidth = 3, linestyle = '-.', label = 'real')
plt.legend()
# plt.grid()
plt.show()
```

численная оценка совпадает с аналитическим значением с точностью $10^{-5}$ на интервале [-1,1]

# Пункт 3:

```python

# Строит матрицу плотности с собственными значениями из распределеня Пуассона
# размера N*N и далее расширяе до размера d*d
def build_density_matrix(N = 5, d = 5, func = poisson_prob, clear = True):
    rho = np.zeros((N,N))
    for n in range(N):
        rho[n][n] = func(n)
        
    rho_full = np.zeros((d,d))
    rho_full[:N,:N] = rho
    return rho_full

# Строит матрицу плотности чистого состояния с собственными значениями из распределеня Пуассона
# размера N*N и далее расширяе до размера d*d
def build_pure_density_matrix(N = 5, d = 5, func = poisson_prob, clear = True):
    rho = np.zeros((N,N))
    psi = np.sqrt(np.array([func(n) for n in range(N)]))
    
    rho = np.outer(psi, psi.conj())
        
    rho_full = np.zeros((d,d))
    rho_full[:N,:N] = rho
    return rho_full

# Строит матрицы операторов рождения\уничтожения
def build_a(N = 5, conj = False):
    a = np.zeros((N,N))
    for n in range(1, N):
        a[n][n-1] = np.sqrt(n)
    if conj == False:
        a = a.T
    return a
```

```python
print('Матрица плотности смешанного состояния: \n \n {}'.format(build_density_matrix()))
print('Матрица плотности чистого состояния: \n \n {}'.format(build_pure_density_matrix()))
```

```python
build_a(conj = False), build_a(conj = True)
```

```python
# Преобразование, осуществляемое светоделителем
def beam_splitter(N = 5):
    a = build_a(N = N, conj = False)
    a_hconj = build_a(N = N, conj = True)
    U = sLA.expm(np.pi/4*(np.kron(a_hconj, a) - np.kron(a, a_hconj)))
    return U

# Квантовая схема и входящие состояния
def circ(rho_in, U = beam_splitter):
    U = beam_splitter(N = int(np.sqrt(rho_in.shape[0])))
    print(U.shape, rho_in.shape)
    rho_out = U @ rho_in @ U.conj().T
    return rho_out
    
```

```python
# Зададим взодящую матрицу плотности согласно заданию (для чистого и смешанного состояний)
N = 5
rho0 = build_density_matrix(N)
rho0p = build_pure_density_matrix(N)
rho1 = np.zeros((N,N))
rho1[0][0] = 1

rho = np.kron(rho0, rho1)
rho_pure = np.kron(rho0p, rho1)


# Получим матрицы плотности выходных состояний для чистого и смешанного входного
rho_out = circ(rho)
rho_out_pure = circ(rho_pure)
```

```python
# Расчитывает вероятности одиночных и совместных кликов фотодетектора для чистого состояния
def get_probs_clear(rho, d = 5):
    p1p2 = np.reshape(np.diag(rho), (d,d))
    p1 = p1p2.sum(axis = 1)
    p2 = p1p2.sum(axis = 0)
    return p1p2, p1, p2

# Расчитывает вероятности одиночных и совместных кликов фотодетектора для произвольного состояния
def get_probs(rho, d = 5):
    rho_tensor = np.reshape(rho, [d,d,d,d])
    p1p2 = np.zeros((d,d))
    p1 = np.zeros(d)
    p2 = np.zeros(d)
    for n1 in range(0, d):
        for n2 in range(0, d):
            p1p2[n1, n2] = np.abs(rho_tensor[n1, n2, n1, n2])
    
    rho_tensor_n1 = np.trace(rho_tensor, axis1=1, axis2=3)
    rho_tensor_n2 = np.trace(rho_tensor, axis1=0, axis2=2)
    
    p1 = np.abs(np.diag(rho_tensor_n1))
    p2 = np.abs(np.diag(rho_tensor_n2))
        
    return p1p2, p1, p2
```

```python
# Получим вероятности кликов для входного и выходных состояний при чистом и смешанном входном

probs_in = get_probs_clear(rho_pure)
probs_out = get_probs(rho_out, 5)
probs_out_pure = get_probs(rho_out_pure, 5)
```

```python
plot_matrix(probs_in[0], title = 'Входное чистое состояние')
plot_matrix(probs_out[0], title = 'Выходное состояние')
plot_matrix(probs_out_pure[0], title = 'Выходное состояние для чистого состояния на входе')
```

```python
plt.title('Сравнение вероятностей возникновения n фотонов в 1-й моде для входа и выхода')
# plt.bar(list(range(N)), probs_in[1], label = 'Входное состояние', alpha = 0.4)
plt.bar(list(range(N)), probs_out[1], label = 'Выходное состояние для смешанного', alpha = 0.4)
plt.bar(list(range(N)), probs_out_pure[1], label = 'Выходное состояние для чистого', alpha = 0.4)
plt.legend()
plt.show()

plt.title('Сравнение вероятностей возникновения n фотонов во 2-й моде для входа и выхода')
# plt.bar(list(range(N)), probs_in[2], label = 'Входное состояние', alpha = 0.4)
plt.bar(list(range(N)), probs_out[2], label = 'Выходное состояние для смешанного', alpha = 0.4)
plt.bar(list(range(N)), probs_out_pure[2], label = 'Выходное состояние для чистого', alpha = 0.4)
plt.legend()
plt.show()
```

```python
# создадим массив матриц, соответствующий проведению серии из num = 100 экспериментов
# в каждом эксперименте проводится M = 1000 запусков схемы и составляется выборка

M = 1000
num = 100

rng = np.random.default_rng()
outclicks12 = rng.multinomial(M, np.ravel(p1p2), size=num).reshape((num,N,N))
outclicks12_pure = rng.multinomial(M, np.ravel(p1p2_pure), size=num).reshape((num,N,N))
outclicks1 = rng.multinomial(M, p1, size=num).reshape((num,N))
outclicks2 = rng.multinomial(M, p2, size=num).reshape((num,N))
outclicks12, outclicks1 
```

```python
# Вычисляет средние n1 , n2 , n1n2
def calc_mean_clicks(outclicks, M = 1000):

    N = outclicks.shape[2]
    n1_clicks = outclicks.sum(axis = 2)
    n2_clicks = outclicks.sum(axis = 1)
    
    weights = np.arange(N)
    weights12 = np.outer(weights, weights)
    
    n1_mean = np.sum(n1_clicks*weights, axis = 1)/M
    n2_mean = np.sum(n2_clicks*weights, axis = 1)/M
    n1n2 = np.sum(outclicks*weights12, axis = (1,2))/M
    
    return n1_mean, n2_mean, n1n2


    
```

```python
# Вычисляет средние n1 , n2 , n1n2
n1_mean, n2_mean, n1n2 = calc_mean_clicks(outclicks12, M)
# На основе средних n1 , n2 , n1n2 вычислим серию g2, среднее значение M(g2) и D(g2)
g2_list = n1n2/(n2_mean*n1_mean)
g2_mean = np.mean(g2_list)
g2_std = np.std(g2_list)
g2_list, g2_mean, g2_std
```

```python
# Аналогично для чистого состояния

n1_mean_pure, n2_mean_pure, n1n2_pure = calc_mean_clicks(outclicks12_pure, M)
g2_list_pure = n1n2_pure/(n2_mean_pure*n1_mean_pure)
g2_mean_pure = np.mean(g2_list_pure)
g2_std_pure = np.std(g2_list_pure)
g2_list_pure, g2_mean_pure, g2_std_pure
```

аналитическое выражение для $g^{(k)}$ записывается как:

$ g^{(k)} = \frac{1}{\mu^{k}}\frac{\partial^{k}}{\partial k^{k}} G(1) $

```python
#Производня
def diff(func, x0 = 1, eps = 1e-2):
    return (func(x0 + eps) - func(x0 - eps))/(2*eps)

#Производная k-го порядка
def diff_k(func, x0 = 1, eps = 1e-2, k = 1):
    if k == 1:
        return diff(func, x0 = x0, eps = eps)
    else:
        return (diff_k(func, x0 = x0 + eps/2, eps = eps/2, k = k - 1) - diff_k(func, x0 = x0 - eps/2, eps = eps/2, k = k - 1))/(eps)
#Значене gk рассчитаное через численные производные    
def g_k(G = G, x0 = 1, k = 2, mu = 0.18):
    return 1/mu**k * diff_k(G, x0 = x0, eps = 1e-1, k = k)
    
```

```python
# Проверка работы функции производной:
def pow3(x):
    return x**3

diff_k(pow3, x0 = 2, eps = 1e-1, k = 3)
```

```python
# Аналитическое значение g2
g_k(x0 = 1, k = 2)
```

Оценка значения g2 совпадает с аналитическим значением в пределах стандартного отклонения

# Пункт 6:

```python
def get_rho_coherent(theta, N, d):
    alpha = 5*np.exp(1j*theta)
    n = np.arange(N)
    fact_n = np.array([np.math.factorial(x) for x in n]).astype(float)
    state = np.exp((-np.abs(alpha) ** 2) / 2) * (alpha ** n) / np.sqrt(fact_n)
    rho = np.outer(state,state.conjugate())
    
    rho_full = np.zeros((d,d))
    rho_full[:N,:N] = rho
    return rho_full
```

```python
def restrictions_err(N, func = poisson_prob):
    return 1 - sum([func(i) for i in range(N)])

def coherent_func(n, alpha = 5):
    return (np.exp((-np.abs(alpha) ** 2) / 2) * (alpha ** n) / np.sqrt(float(np.math.factorial(n))))**2
```

Рассчитаем размерность когерентного состояния, необходимую для ошибки $err < 10^{-5}$

```python
print('Ошибка в распределении за счёт сокращения размерности до 4:', restrictions_err(N = 49, func = coherent_func))
print('Ошибка в распределении за счёт сокращения размерности до 5:', restrictions_err(N = 50, func = coherent_func))
```

Размерность когерентного состоянии должна быть не менее 50
